```python
import numpy as np
def cosine_similarity(a,b):
    dot_prod = np.dot(a,b)

    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    print(dot_prod/(norm_a*norm_b))

king_vec = model.get_vector("king")
queen_vec = model.get_vector("queen")
```
